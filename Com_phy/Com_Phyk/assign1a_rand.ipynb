```python
#____QUESTION_1____________

import matplotlib.pyplot as plt
import numpy as np
import Adilib as adi

def index_f(n):
    return [(i+1)/n for i in range(n)]
    
def Plot(x, y , title='Sample Plot', xlabel='X-axis Label', ylabel='Y-axis Label',file_name='sample_plot.png'):
    plt.plot(x, y, marker='o', linestyle='none', color='b')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)  
    plt.savefig(file_name)
    plt.show()

def pseudo1(n,c,s=0.1):
    x_i = s
    l=[]
    for _ in range(n):
        x_i = c*x_i*(1-x_i)
        l.append(x_i)
    return l

def corre_pseudo1(n,c,k=5):
    result=[]
    a=pseudo1(n,c)[:-k]
    b=pseudo1(n,c)[k:]
    result.append(a)
    result.append(b)
    return result


list_of_different_c_values=[3.857318484034,8.9762561527,3.79936826586586]
list_of_different_k_values=[5,10,15]


N=1000
for _ in list_of_different_c_values:
    Plot(index_f(N),pseudo1(N,_),'c=3.857318484034, x0=0.1')
    for __ in list_of_different_k_values:
        a=corre_pseudo1(N,_)[0]
        b=corre_pseudo1(N,_)[1]
        Plot(a,b,'k=5,c=3.857318484034')
```

```python
#_______QUESTION_2__________________

import matplotlib.pyplot as plt
import numpy as np
import Adilib as adi

def index_f(n):
    return [(i+1)/n for i in range(n)]

def Plot(x, y , title='Sample Plot', xlabel='X-axis Label', ylabel='Y-axis Label',file_name='sample_plot.png'):
    plt.plot(x, y, marker='o', linestyle='none', color='b')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)  
    plt.savefig(file_name)
    plt.show()
    
def LCG(N,a=1103515245,c=12345,m=32768,x_0=.1):
    l=[]
    x_i=x_0
    for i in range(N):
        x_i= ((a*x_i +c)%m)
        l.append(x_i/m)
    return l


def corre_LCG(N,k):
    result=[]
    a=LCG(N)[:-k]
    b=LCG(N)[k:]
    result.append(a)
    result.append(b)
    return result

N=1000
k=5
a=corre_LCG(N,k)[0]
b=corre_LCG(N,k)[1]
Plot(index_f(N),LCG(N),file_name='LCG(N=1000)')
Plot(a,b,file_name='Correlation_LCG')
```
