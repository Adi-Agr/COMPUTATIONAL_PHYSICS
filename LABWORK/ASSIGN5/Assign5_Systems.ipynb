```python
import sys
import numpy as np
import matplotlib.pyplot as plt

sys.path.append('C:/Users/agraw/Documents/GitHub/COMPUTATIONAL_PHYSICS/LABWORK')

import adilib_main as ad
```

```python
# Define the 6x6 system matrix A and vector b from the assignment
A1 = [
    [4, -1, 0, -1, 0, 0],
    [-1, 4, -1, 0, -1, 0],
    [0, -1, 4, 0, 0, -1],
    [-1, 0, 0, 4, -1, 0],
    [0, -1, 0, -1, 4, -1],
    [0, 0, -1, 0, -1, 4]
]

b1 = [2, 1, 2, 2, 1, 2]

print("Matrix A1:")
for row in A1:
    print(row)
print("\nVector b1:", b1)
```

```python
print("\nIs A1 symmetric?-->", ad.is_symmetric(A1))
```

```python
# Solve the first system using Cholesky decomposition
print("\n--- Solving System 1 using Cholesky factorization ---")
x_cholesky = ad.solve_by_cholesky(A1, b1)
print("\nSolution x (Cholesky):", [f"{val:.6f}" for val in x_cholesky])
```

```python
# Solve the first system using Gauss-Seidel
print("\n--- Solving System 1 using Gauss-Seidel Iteration ---")
x_gauss_seidel, iterations = ad.gauss_seidel_iteration(A1, b1, tol=1e-6)
print("\nSolution x (Gauss-Seidel):", [f"{val:.6f}" for val in x_gauss_seidel])
```

```python
# Define the 5x5 system matrix A and vector b from the assignment
A2 = [
    [4, 0, 4, 10, 1],
    [0, 4, 2, 0, 1],
    [2, 5, 1, 3, 13],
    [11, 3, 0, 1, 2],
    [3, 2, 7, 1, 0]
]

b2 = [20, 15, 92, 51, 15]

print("A2:")
for row in A2:
    print(row)
print("\nb2:", b2)


#MANUALLY MAKING IT DIAGONALLY DOMINANT
A3 = [
    [11,3,0,1,2],
    [0,4,2,0,1],    
    [3,2,7,1,0],
    [4,0,4,10,1],
    [2,5,1,3,13],
]
b3 = [51,15,15,20,92]
print()
print("A3:")
for row in A2:
    print(row)
print("\nb3:", b2)
```

```python
# Solve the rearranged system using Jacobi method
print("\n--- Solving Rearranged System 2 using Jacobi Iteration ---")
x_jacobi,iterations_jacobi = ad.jacobi_iteration(A3,b3, tol=1e-6)
print("\nSolution x (Jacobi - rearranged):", [f"{val:.6f}" for val in x_jacobi])
```
