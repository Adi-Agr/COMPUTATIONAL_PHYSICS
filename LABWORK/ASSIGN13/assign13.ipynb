```python
import sys
import matplotlib.pyplot as plt

sys.path.append('c:/Users/mayur/ClonedGitRepo/COMPUTATIONAL_PHYSICS/LABWORK')

import adilib_class as ad
import math
```

```python
#_______QUESTION_1______________
f = lambda x,y: y - x**2
x0=0
y0=0
xf=2
dx=0.1
x_euler, y_euler = ad.ODEs.euler_solver(f,y0,x0,xf,dx)
x_peco, y_peco = ad.ODEs.predictor_corrector_solver(f,y0,x0,xf,dx)
f_analytic = lambda x: (x**2 + 2*x + 2 - 2*math.exp(x))
plt.plot(x_euler, y_euler, label='Euler Method', color='blue')
plt.plot(x_peco, y_peco, label='Pe-Co Method', color='green')
plt.plot(x_euler, [f_analytic(x) for x in x_euler], label='Analytic Solution', color='red', linestyle='dashed')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solving ODE dy/dx = y - x^2')
plt.legend()
plt.grid()
plt.show()
```

```python
#_______QUESTION_2______________
f = lambda x,y: (x + y)**2
x0=0
y0=1
xf=math.pi/5
dx=0.1
x_euler, y_euler = ad.ODEs.euler_solver(f, y0, x0, xf, dx)
x_peco, y_peco = ad.ODEs.predictor_corrector_solver(f, y0, x0, xf, dx)
f_analytic = lambda x: (math.tan(x+ math.pi/4) - x)
plt.plot(x_euler, y_euler, label='Euler Method', color='blue')
plt.plot(x_peco, y_peco, label='Pe-Co Method', color='green')
plt.plot(x_euler, [f_analytic(x) for x in x_euler], label='Analytic Solution', color='red', linestyle='dashed')
plt.title('Solving ODE dy/dx = (x + y)^2')
plt.grid()
plt.legend()
plt.show()
```
